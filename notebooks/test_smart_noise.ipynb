# SmartNoise Synthesizer Test
This notebook validates the `SmartNoiseSynthesizer` implementation.

```python
import os
import sys
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Add project root to path
sys.path.append(os.path.abspath(os.path.join(os.path.dirname(''), '..')))

from src.synthesizing.smart_noise import SmartNoiseSynthesizer
from src.entities.dataset import Dataset
from src.entities.denial_constraints import DenialConstraints
```

## 1. Setup Data

```python
data = pd.DataFrame({
    'age': np.random.randint(18, 80, 1000),
    'income': np.random.randint(20, 150, 1000),
    'gender': np.random.choice(['M', 'F'], 1000)
})
dcs = DenialConstraints([])
ds = Dataset("adult_mini", data, dcs, "income")
print(f"Original data shape: {data.shape}")
```

## 2. PATECTGAN Test

```python
%%time
synth_pate = SmartNoiseSynthesizer(engine="patectgan", epsilon=1.0)
res_pate = synth_pate.synthesize(ds)
print(f"PATECTGAN data shape: {res_pate.data.shape}")
```

## 3. Visualization

```python
fig, ax = plt.subplots(1, 2, figsize=(12, 5))
data['age'].hist(ax=ax[0], alpha=0.5, label='Original')
res_pate.data['age'].hist(ax=ax[0], alpha=0.5, label='Synthetic')
ax[0].set_title('Age Distribution')
ax[0].legend()

data['gender'].value_counts().plot(kind='bar', ax=ax[1], alpha=0.5, label='Original', position=0, width=0.4)
res_pate.data['gender'].value_counts().plot(kind='bar', ax=ax[1], alpha=0.5, color='orange', label='Synthetic', position=1, width=0.4)
ax[1].set_title('Gender Distribution')
ax[1].legend()
plt.show()
```

## 4. Logical Tests

```python
assert res_pate.data.shape == data.shape, "Shape mismatch!"
assert list(res_pate.data.columns) == list(data.columns), "Column mismatch!"
print("All logical tests passed!")
```
